```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Assuming your dataset is stored in a DataFrame called 'data'
# Extract features and target variable
X = df_meta.drop(['user_snippet', 'activity'], axis=1)  # Remove non-feature columns
y = df_meta['activity']

# Initialize and fit the RandomForestClassifier model
model = RandomForestClassifier()
model.fit(X, y)

# Get feature importances
feature_importances = model.feature_importances_

# Create a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort the DataFrame by feature importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the top 10 most important features
print(feature_importance_df)
```

```python
# Calculate the mean importance value
mean_importance = feature_importance_df['Importance'].mean()

# Filter the DataFrame to keep only features with importance scores greater than or equal to the mean
selected_features_df = feature_importance_df[feature_importance_df['Importance'] >= mean_importance]

# Extract the names of the remaining features
selected_feature_names = selected_features_df['Feature'].tolist()

# Drop features from the original DataFrame that are not in the selected list of feature names
X_selected = df_meta[selected_feature_names]

# Now X_selected contains only the selected features for further analysis or modeling
```

```python
X_selected.info()
```

```python
# Drop the same features from the test set as from the training set
X_test_selected = df_test[selected_feature_names]
y_test_df=df_test['activity']
# Now X_test_selected contains only the selected features from the test set
```

```python

# Initialize and fit the RandomForestClassifier model
model = RandomForestClassifier()
model.fit(X_selected, y)
```

```python
y_test_pred=model.predict(X_test_selected)
# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test_df, y_test_pred)
print("Test Accuracy:", test_accuracy)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Define your RandomForestClassifier model with desired hyperparameters
model = RandomForestClassifier()

# Perform cross-validation with 5 folds
cv_scores = cross_val_score(model, X_selected, y, cv=10)

# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)

# Calculate the mean cross-validation score
mean_cv_score = cv_scores.mean()
print("Mean Cross-validation score:", mean_cv_score)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define the hyperparameters grid to search
param_grid = {
    'n_estimators': [50, 100, 200,300],
    'max_depth': [None, 10, 20,30],
    'min_samples_split': [2, 5, 10,15],
    'min_samples_leaf': [1, 2, 4,6]
}

# Initialize RandomForestClassifier
rf_classifier = RandomForestClassifier()

# Initialize GridSearchCV with the RandomForestClassifier model and hyperparameter grid
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=10, scoring='accuracy')

# Perform grid search to find the best hyperparameters
grid_search.fit(X_selected, y)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best cross-validation score
best_score = grid_search.best_score_
print("Best Cross-validation Score:", best_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize RandomForestClassifier with the best hyperparameters
best_rf_classifier = RandomForestClassifier(max_depth=18, min_samples_leaf=2, min_samples_split=4, n_estimators=200)

# Train the model on the entire training data
best_rf_classifier.fit(X_selected, y)

# Make predictions on the test set
y_test_pred = best_rf_classifier.predict(X_test_selected)

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test_df, y_test_pred)
print("Test Accuracy with Best Hyperparameters:", test_accuracy)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint

# Define the hyperparameters distribution to search
param_dist = {
    'n_estimators': randint(50, 500),
    'max_depth': [None] + list(range(5, 30)),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

# Initialize RandomForestClassifier
rf_classifier = RandomForestClassifier()

# Initialize RandomizedSearchCV with the RandomForestClassifier model and hyperparameter distribution
random_search = RandomizedSearchCV(estimator=rf_classifier, param_distributions=param_dist, n_iter=100, cv=10, scoring='accuracy', random_state=42)

# Perform random search to find the best hyperparameters
random_search.fit(X_selected, y)

# Get the best hyperparameters
best_params_random = random_search.best_params_
print("Best Hyperparameters (Random Search):", best_params_random)

# Get the best cross-validation score
best_score_random = random_search.best_score_
print("Best Cross-validation Score (Random Search):", best_score_random)
```
